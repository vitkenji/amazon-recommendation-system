```python
import pandas as pd
import networkx as nx
import numpy as np
import collections
import ast
```

```python
G = nx.read_gml('../network/reviews_network.gml')
```

```python
df = pd.read_csv('../dataset/new_features.csv')
```

```python
df = df.drop(columns=['Unnamed: 0', 'rating', 'neg', 'neu', 'pos', 'compound'], axis=1)
```

```python
df["categories"] = df["categories"].apply(ast.literal_eval)
df["number_categories"] = df["categories"].apply(len)
```

```python
df.head()
```

```python
df["exists"] = 1

users = df["user_id"].unique()
products = df["parent_asin"].unique()

n_positive = len(df)

existing_edges = set(zip(df["user_id"], df["parent_asin"]))

neg_samples = []
rng = np.random.default_rng()

while len(neg_samples) < n_positive:
    u = rng.choice(users)
    p = rng.choice(products)

    if (u, p) not in existing_edges:
        neg_samples.append((u, p))

df_neg = pd.DataFrame(neg_samples, columns=["user_id", "parent_asin"])
df_neg["exists"] = 0

df = pd.concat([df, df_neg], ignore_index=True)
```
